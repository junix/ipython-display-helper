```python
def _attr2(o, attr_name):
    if callable(attr_name):
        v = attr_name(o)
        return v
    
    
    if not hasattr(o, attr_name):
        return '×'
        
    v = getattr(o, attr_name,None)
    if attr_name in ('__class__',):
        return v
    if callable(v):
        try:
            return v()
        except:
            return type(v).__name__
    return v if v is not None else '-'

def _attr(o, attr_name):
    v = _attr2(o,attr_name)
    if isinstance(v, str) or v is None:
        return v
    
    vs = str(v)
    if len(vs) > 12:
        vs = vs[:12]+' ...'
    if vs.startswith('<class'):
        return v.__name__
    
    if isinstance(v, bool|str|dict|tuple|list):
        return v
    
    
    return vs + f" ◀ {type(v).__name__}"
```

```python
def compare(tys, ms=ms):
    return pd.DataFrame(
        {
            k: [_attr(v, e) for e in ms] for k, v in tys.items()
        },
        index=[getattr(x, '__name__', x) for x in ms]
    )
```

```python
import io
from pygments import highlight
from pygments import lexers
from pygments.formatters.html import HtmlFormatter
from pygments.formatters.terminal256 import Terminal256Formatter
from pygments.styles import get_style_by_name,get_all_styles
from IPython import display
from IPython.core.display import HTML

def show_source(obj, style_name="algol"):
    # Settings
    lexer = lexers.get_lexer_by_name('python')
    style = get_style_by_name(style_name)
    formatter = HtmlFormatter(full=True, style=style)

    # Highlight
    html = HTML(highlight(inspect.getsource(obj), lexer, formatter))
    display.display_html(html)
    
def show_source2(obj, style_name="algol"):
    # Settings
    lexer = lexers.get_lexer_by_name('python')
    style = get_style_by_name(style_name)
    formatter = Terminal256Formatter(style=style)
    print(highlight(inspect.getsource(obj), lexer, formatter))
```
