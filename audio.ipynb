````python
import pyaudio
import wave

# 在这个代码中，PyAudio 库初始化了一个录音流，以 44100Hz 的采样频率、2 个声道、16bit 量化的 paInt16 格式采样，
# 每次采样 1024 个样本，录音 3 秒钟。最后，采样的数据存储到了一个 WAV 文件中。

chunk = 102
sample_format = pyaudio.paInt16
channels = 1
fs = 44100
seconds = 10
filename = "output.wav"

p = pyaudio.PyAudio()

stream = p.open(format=sample_format,
                channels=channels,
                rate=fs,
                frames_per_buffer=chunk,
                input=True)

frames = []

for i in range(0, int(fs / chunk * seconds)):
    data = stream.read(chunk)
    frames.append(data)

stream.stop_stream()
stream.close()

p.terminate()

wf = wave.open(filename, 'wb')
wf.setnchannels(channels)
wf.setsampwidth(p.get_sample_size(sample_format))
wf.setframerate(fs)
wf.writeframes(b''.join(frames))
wf.close()
````

#### [#](https://www.xfyun.cn/doc/asr/rtasr/API.html#signa生成)signa生成

1.获取baseString，baseString由appid和当前时间戳ts拼接而成，假如appid为595f23df，ts为1512041814，则baseString为

> 595f23df1512041814

2.对baseString进行MD5，假如baseString为上一步生成的595f23df1512041814，MD5之后则为

> 0829d4012497c14a30e7e72aeebe565e

3.以apiKey为key对MD5之后的baseString进行HmacSHA1加密，然后再对加密后的字符串进行base64编码。
假如apiKey为d9f4aa7ea6d94faca62cd88a28fd5234，MD5之后的baseString为上一步生成的0829d4012497c14a30e7e72aeebe565e，
则加密之后再进行base64编码得到的signa为

> IrrzsJeOFk1NGfJHW6SkHUoN9CU=

备注：

- apiKey：接口密钥，在应用中添加实时语音转写服务时自动生成，调用方注意保管；
- signa的生成公式：HmacSHA1(MD5(appid + ts), api_key)，具体的生成方法详见【[调用示例](https://www.xfyun.cn/doc/asr/rtasr/API.html#调用示例)】；

#### [#](https://www.xfyun.cn/doc/asr/rtasr/API.html#请求示例)请求示例

```text
	ws://rtasr.xfyun.cn/v1/ws?appid=595f23df&ts=1512041814&signa=IrrzsJeOFk1NGfJHW6SkHUoN9CU=&pd=edu
```

#### [#](https://www.xfyun.cn/doc/asr/rtasr/API.html#返回值)

````python
from hashlib import md5, sha1
import hmac
import base64
import urllib


def sign(*, app_id: str, ts: str, api_key: str):
    base_str = app_id + ts
    x = md5()
    x.update(base_str.encode('utf-8'))
    base_str = x.hexdigest()

    hmac_code = hmac.new(api_key.encode(), base_str.encode(), sha1)
    code = hmac_code.digest()
    return base64.b64encode(code).decode('utf-8')


def make_params(*, app_id: str, api_key: str):
    ts = str(int(time.time()))
    params = urllib.parse.urlencode({
        'appid': app_id, 
        'ts':ts, 
        'signa': sign(app_id=app_id,ts=ts,api_key=api_key)}
    )
    return f'wss://rtasr.xfyun.cn/v1/ws?{params}'
````
